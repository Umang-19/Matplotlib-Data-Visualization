```python
# Importing Libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for plotiing Line Plot

xvalues = np.array([0,10])
yvalues = np.array([0,100])
```

```python
# Plotting Line Plot

plt.plot(xvalues,yvalues)
plt.show()
```

```python
# Example 2

x = np.arange(1,51)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
y = -2*x
y
```

```python
# Setting title, xlabel and ylabel

plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Adding some styling

plt.plot(x,y, color='green', linewidth=5, linestyle=":", marker='o')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Plotting multiple line plots with grid
x1 = np.arange(1,11)
y1 = 2*x1 + 5
y2 = 3*x1 + 4

plt.plot(x1,y1, color='green',linestyle='dashed',linewidth=2, marker='o')
plt.plot(x1,y2,color='red', linewidth=2, marker='*')
plt.grid(True, color='black')
plt.show()
```
