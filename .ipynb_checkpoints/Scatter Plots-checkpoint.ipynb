```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
y1 = x**4
y2 = x**3 + 7
```

```python
plt.style.use('seaborn')
plt.figure(figsize = (6,6))
plt.scatter(x,y1, color='red', label='Plot 1')
plt.scatter(x,y2, color='yellow', label='Plot 2')
plt.legend()
plt.show()
```
