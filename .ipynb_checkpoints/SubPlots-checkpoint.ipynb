```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = -3*x

# subplot(no. of rows, no. of columns, plot index)
plt.subplot(1,2,1)
plt.plot(x,y1, linewidth=2, color='red')
plt.grid(True)
plt.title('Plot 1')

plt.subplot(1,2,2)
plt.plot(x,y2, linestyle=':', linewidth=3)
plt.grid(True)

plt.title('Plot 2')
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1, linewidth=2, color='red')
plt.grid(True)
plt.title('Plot 1')

plt.subplot(2,1,2)
plt.plot(x,y2, linestyle=':', linewidth=3)
plt.grid(True)

plt.title('Plot 2')
plt.show()
```
